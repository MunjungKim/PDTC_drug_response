```python
import pandas as pd
```

# preprocess tcga data

```python
response_tcga = pd.read_csv('../data/clinical_table.csv')
```

```python
response_tcga
```

```python
response = response_tcga[['bcr_patient_barcode', 'pharmaceutical_therapy_drug_name', 'treatment_best_response']]
```

```python
response
```

```python
response_type = ['Complete Response', 'Partial Response', 'Stable Disease', 'Clinical Progressive Disease']
```

```python
response_exist = response.loc[response['treatment_best_response'].isin(response_type)]
response_exist = response_exist.reset_index(drop=True)
```

```python
response_exist
```

```python
drug_correction_dic = {}
file = open('../data/drug_synonym_match.txt', 'r')
for line in file:
    line = line.rstrip().split('\t')
    drug_correction_dic[line[0]] = line[1]
```

```python
drug_correction_dic
```

```python
response_exist_conversion = pd.DataFrame()
for idx, line in response_exist.iterrows():
    if line[1] in drug_correction_dic.keys():
        temp = pd.DataFrame({'bcr_patient_barcode' : [line[0]], 'pharmaceutical_therapy_drug_name' : [drug_correction_dic[line[1]]], 'treatment_best_response' : [line[2]]})
        response_exist_conversion = response_exist_conversion.append(temp, ignore_index=True)
```

```python
response_exist_conversion
```

```python
response_exist_conversion.to_csv('../data/tcga_response_exist_conversion.csv', sep='\t')
```

# prepare fingerprint conversion

```python
tcga_drug_set = set(response_exist_conversion['pharmaceutical_therapy_drug_name'].values)
```

```python
tcga_drug_set
```

```python
with open('../data/tcga_drug_list.txt', 'w') as f:
    for item in tcga_drug_set:
        f.write('%s\n' % item)
```

# gene name conversion

```python
tcga_patient_set = set(response_exist_conversion['bcr_patient_barcode'].values)
```

```python
expression_tcga = pd.read_csv('../data/TMM_log2_rna_seq.txt', sep='\t')
```

```python
expression_tcga
```

```python
have_response_index = []
for i, index in enumerate(expression_tcga.columns):
    if index[:12] in tcga_patient_set:
        have_response_index.append(i)
have_response_index = [0] + have_response_index
```

```python
have_response_index
```

```python
expression_tcga_have_response = expression_tcga.iloc[:,have_response_index]
```

```python
expression_tcga_have_response
```

```python
file = open('../data/gene_symbol.txt', 'r')
gene_symbol_dic = {}
for line in file:
    line = line.rstrip().split('\t')
    if len(line) > 1 and line[1] != 'HGNC symbol':
        gene_symbol_dic[line[0]] = line[1]
```

```python
gene_symbol_dic
```

```python
expression_tcga_have_response_conversion = expression_tcga_have_response.loc[expression_tcga_have_response['genes'].isin(gene_symbol_dic.keys())]
expression_tcga_have_response_conversion = expression_tcga_have_response_conversion.reset_index(drop=True)
```

```python
expression_tcga_have_response_conversion
```

```python
for idx in range(len(expression_tcga_have_response_conversion)):
    ensmbl = expression_tcga_have_response_conversion.iloc[idx, 0]
    expression_tcga_have_response_conversion.iloc[idx, 0] = gene_symbol_dic[ensmbl]
```

```python
expression_tcga_have_response_conversion
```

```python
expression_tcga_have_response_conversion.to_csv('../data/')
```
