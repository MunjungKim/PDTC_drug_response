# Ridge

```python
alphas = 10**np.linspace(5,-4,81)
```

```python
rg_grid_pdtc = GridSearchCV(Ridge(normalize=True), cv=5, param_grid={'alpha':alphas}, iid=False)
lasso_grid_pdtc = GridSearchCV(Lasso(normalize=True), cv=5, param_grid={'alpha':alphas}, iid=False)
```

```python
def making_ridge(x,y):
    rg_grid = GridSearchCV(Ridge(normalize=True), cv=5, param_grid={'alpha':alphas}, iid=False)
    X_ccle_train, X_ccle_test, Y_ccle_train, Y_ccle_test = train_test_split(x,y, test_size=0.15)
    rg_grid.fit(X_ccle_train,Y_ccle_train)
    alpha = rg_grid.best_params_
    score = rg_grid.score(X_ccle_test,Y_ccle_test)
    
    return rg_grid,alpha,score
    
```

```python
def making_lasso(x,y):
    lasso_grid = GridSearchCV(Lasso(normalize=True), cv=5, param_grid={'alpha':alphas}, iid=False)
    X_ccle_train, X_ccle_test, Y_ccle_train, Y_ccle_test = train_test_split(x,y, test_size=0.15)
    lasso_grid.fit(X_ccle_train,Y_ccle_train)
    alpha = lasso_grid.best_params_
    score = lasso_grid.score(X_ccle_test,Y_ccle_test)
    
    return lasso_grid,alpha,score
```

## CCLE multiple model

```python
rg_list = []
lasso_list = []
#model list
start = time.time()
for i in tqdm(range(len(sampleList))):
    rg_list.append([making_ridge(X_Cell[sampleList[i]],Y_Cell[sampleList[i]])[0]])
    lasso_list.append([making_lasso(X_Cell[sampleList[i]],Y_Cell[sampleList[i]])[0]])
    break
end = time.time()

print(end-start)
```
