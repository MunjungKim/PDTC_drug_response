## L1000

```python
FILE_L1000_PATH =  '../data/landmark.txt'
l1000 = pd.read_csv(FILE_L1000_PATH, sep = "\t",  engine='python', encoding = "cp949")
```

```python
l1000
```

```python
l1000_symbol = l1000['Symbol']
```

```python
l1000_symbol
```

### Extracting : pdtc

```python
exp_model_l1000 = exp_model[exp_model.index.isin(l1000_symbol)]
```

```python
exp_model_l1000
```

### Extracting : cell line

```python
cell_line_l1000 = cell_line[cell_line.index.isin(l1000_symbol)]
```

```python
cell_line_l1000
```

### Extracting : tcga

```python
tcga_l1000 = tcga[tcga.index.isin(l1000_symbol)]
```

```python
tcga_l1000
```

```python
for i in range(len(exp_model_l1000.columns)):
    exp_model_l1000.iloc[:, i] = exp_model_l1000.iloc[:, i].rank(method='min', ascending=True) / len(l1000_symbol)
for i in range(len(cell_line_l1000.columns)):
    cell_line_l1000.iloc[:, i] = cell_line_l1000.iloc[:, i].rank(method='min', ascending=True) / len(l1000_symbol)
for i in range(len(tcga_l1000.columns)):
    tcga_l1000.iloc[:, i] = tcga_l1000.iloc[:, i].rank(method='min', ascending=True) / len(l1000_symbol)
```

```python
# file save
PDTC_L1000_SAVE = '../data/with_human/pdtc_L1000_rank.csv'
exp_model_l1000.to_csv(PDTC_L1000_SAVE, sep=',', na_rep='NaN') 

CELL_L1000_SAVE = '../data/with_human/cell_line_L1000_rank.csv'
cell_line_l1000.to_csv(CELL_L1000_SAVE, sep=',', na_rep='NaN') 

TCGA_SAVE = '../data/with_human/tcga_L1000_rank.csv'
tcga_l1000.to_csv(TCGA_SAVE, sep=',', na_rep='NaN') 

```
